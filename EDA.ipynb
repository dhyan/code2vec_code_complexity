```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
# Reading the csv file and putting it into 'df' object.
df = pd.read_csv('metadata.csv')
df.head()
```

```python
print(df.shape)
print(df.columns)
```

```python
# df.nunique(axis=0)
df.describe().apply(lambda s: s.apply(lambda x: format(x, 'f')))
```

```python
# calculate correlation matrix
data = df.corr()# plot the heatmap
f, ax = plt.subplots(figsize=(30, 25))
sns.heatmap(data, 
        linewidths=0.1, annot=True, annot_kws={"size":15},cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

```python
sns.pairplot(df)
```

```python
df_complexity = df.groupby('complexity').size()
labels = ['1', 'logn', 'n', 'n_square', 'nlogn']
sizes = [df_complexity['1'], df_complexity['logn'], df_complexity['n'], df_complexity['n_square'], df_complexity['nlogn']]
plt.figure(figsize=(15,10))
explode = (0, 0.1, 0, 0, 0)
plt.pie(sizes, labels=labels, explode=explode, shadow=True, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Different complexity classses present in dataset')
plt.show()
```
