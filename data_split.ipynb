```python
from pathlib import Path
import shutil
import pandas as pd
```

```python
# Reading the csv file and putting it into 'df' object.
df = pd.read_csv('metadata.csv')
df.head()
```

```python
def find_complexity(file_name):
    complexity = df.loc[df['file_name'] == file_name].complexity
    complexity = complexity.tolist()[-1]
    return complexity
```

```python
base_path = Path.cwd()
base_path
```

```python
output_path = base_path.joinpath('dataset_split')
output_path
```

```python
for dir in output_path.iterdir():
    print(dir)
```

```python
data_path = base_path.joinpath('Dataset')
data_path
```

```python
for java_file in data_path.iterdir():
    print(str(java_file).split('/')[-1])
    file_name = str(java_file).split('/')[-1]
    class_name = find_complexity(file_name)  # which class directory to copy to
    data_class_path = output_path.joinpath(class_name)
    print(data_class_path)
    shutil.copy(java_file, data_class_path)
```
